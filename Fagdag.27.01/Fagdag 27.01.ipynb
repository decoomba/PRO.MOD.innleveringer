```python
from pylab import *            # 7.6

def planter(t):
    N = 0.2*2.771**t + 90
    return N
dx = 1E-10
def derivert(planter, t, dx):
    fder = (planter(t+dx)-planter(t))/dx
    return fder

tid = []
plant = []
endr = []

for i in range(0,11):
    antall = planter(i)
    tid.append(i)
    plant.append(antall)
    endring = derivert(planter, i, dx)
    endr.append(endring)
    
    
plot(tid, plant)
plot(tid, endr)

    
```

```python
from pylab import *                              # 7.9

dx = 1E-10
tempratur = []
end_temp = []
tid = []

x_verdier = linspace(0, 60, 1000)

def Temp(t):
    T = 70*2.718**(-0.065*t)
    return T

def derivert(Temp, t, dx):
    fder = (Temp(t+dx)-Temp(t))/dx
    return fder

for x in x_verdier:
    C = Temp(x)
    tempratur.append(C)
    c = derivert(Temp, x, dx)
    end_temp.append(c)
    tid.append(x)
    
plot(x_verdier, tempratur)
plot(x_verdier, end_temp)


print(end_temp[41])

    
    
```

```python
from pylab import *

data = loadtxt('posisjon.txt', skiprows=1, delimiter='        ')

tid = data[:,0]
posisjon = data[:,1]

fart = []
for i in range(len(tid)-1):
    dy = posisjon[i+1]-posisjon[i]
    dx = tid[i+1]-tid[i]
    derivert = dy/dx
    fart.append(derivert)
    
akslerasjon = []
for o in range(len(fart)-1):
    dY = fart[o+1]-fart[o]
    dX = tid[o+1]-tid[o]
    derivert1 = dY/dX
    akslerasjon.append(derivert1)
    
plot(tid[:-2],akslerasjon, label = "akslerasjon")
plot(tid[:-1], fart, label = "fart")
plot(tid, posisjon, label = "posisjon")
legend()
show()
```
